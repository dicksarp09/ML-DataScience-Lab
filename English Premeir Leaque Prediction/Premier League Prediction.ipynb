```python
df["Hour"] = df["Time"].str.replace(r":.+", "", regex=True).astype(int)
```

```python
# Create a new column combining HomeTeam and AwayTeam
df['Venue'] = df['HomeTeam'].astype("category").cat.codes

```

```python
def determine_winner(row):
    if row['FT Home TeamG'] > row['FT Away TeamG']:
        return row['HomeTeam']
    elif row['FT Home TeamG'] < row['FT Away TeamG']:
        return row['AwayTeam']
    else:
        return 'Draw'

df['Winner'] = df.apply(determine_winner, axis=1)

```

```python
def encode_winner(row):
    if row['FT Home TeamG'] > row['FT Away TeamG']:
        return 1  # HomeTeam wins
    else:
        return 0  # AwayTeam wins or Draw

# Apply the function to create the new column
df['Outcome'] = df.apply(encode_winner, axis=1)
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("Date")
    rolling_stats = group[cols].rolling(3,closed = "left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["Home Team Shots","HT Home TeamG","HT Away TeamG","Away Team Shots","Home Short Target","Away Short Target","HF", "AF","Home Corners","Away Corners"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = df.groupby("HomeTeam").apply(lambda x : rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('HomeTeam')
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["Date"] < '2020-10-10']
    test = data[data["Date"] >= '2020-10-10']
    rf.fit(train[predictors], train["Outcome"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["Outcome"], predicted=preds), index=test.index)
    precision = precision_score(test["Outcome"], preds)  
    return combined, precision                   
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
